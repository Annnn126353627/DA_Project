## Plot for Dataset

```python
df.head()
```

```python
df['Year'] = pd.DatetimeIndex(df['Date/Time']).year
df['Month'] = pd.DatetimeIndex(df['Date/Time']).month
df
```

```python
df['Temp_C'].unique()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(9, 5))
colors = np.random.randint(8784, size=(8784))
g1 = sns.scatterplot(data=df, x = 'Temp_C', y = 'Rel Hum_%', ax=ax[0], alpha=0.4, hue='Month')
g1.set_title('Relationship temperature and humidity')
g2 = sns.scatterplot(data=df, x = 'Dew Point Temp_C', y = 'Rel Hum_%', ax=ax[1], alpha=0.4, hue='Month')
g2.set_title('Relationship dew point temperature and humidity');
```

```python
df.head(2)
```

```python
df = df.rename(columns={'Weather Condition': 'Weather_Condition'})
split_weather = df.set_index('Year').Weather_Condition.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
plt.figure(figsize=(9, 5))
sns.countplot(y=split_weather, order=split_weather.value_counts().index[:20])
plt.title('Top 20 popular weather');
```

```python
# Plot for Rel Hum %
total_Rel_hum = df.groupby('Month')['Rel Hum_%'].sum()
mean_of_Rel_hum = df.groupby('Month')['Rel Hum_%'].mean()
```

```python
total_Rel_hum.plot(kind='bar', x='Year', y='Rel Hum_%')
plt.title('Sum of Rel Hum in the Month')
plt.grid(alpha=0.07)
```

```python
mean_of_Rel_hum.plot(kind='bar', x='Year', y='Rel Hum_%')
plt.title('Mean of Rel Hum in the Month')
plt.grid(alpha=0.07)
```

```python
#Mean of Temp C and Dev Point Temp C
mean_of_Temp_C = df.groupby('Month')['Temp_C'].mean()
mean_of_Dew_Point_Temp_C = df.groupby('Month')['Dew Point Temp_C'].mean()
print('{}\n\n{}'.format(mean_of_Temp_C, mean_of_Dew_Point_Temp_C))
```

```python
color = np.random.randint(12, size=(12))
mean_of_Temp_C.plot(kind='bar',  x='Month', y='Temp_C', color = 'darkred')
plt.title('Mean of Temp_C');
#mean_of_Dew_Point_Temp_C.plot(kind='bar', x='Month', y='Dew Point Temp_C'
```

```python
mean_of_Dew_Point_Temp_C.plot(kind='bar', x='Month', y='Dew Point Temp_C')
plt.title('Mean of Dew Point Temp_C');
```
