```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data
pl16 = pd.read_csv('pl_15-16.csv')
pl17 = pd.read_csv('pl_16-17.csv')
pl18 = pd.read_csv('pl_17-18.csv')
pl19 = pd.read_csv('pl_18-19.csv')
pl20 = pd.read_csv('pl_19-20.csv')
```

```python
#add collumns year for data

pl16['year'] = '2015/16'
pl17['year'] = '2016/17'
pl18['year'] = '2017/18'
pl19['year'] = '2018/19'
pl20['year'] = '2019/20'
```

```python
pl_upto_20 = pd.concat([pl16, pl17, pl18, pl19, pl20])
pl_upto_20
```

```python
pl_upto_20.info()
```

```python
# TOP 10 SCRORE IN 2015/16
pl16 = pl16[pl16['Appearances'] != 0]
goalscore = pl16.nlargest(10, 'Goals')
goalscore.plot(kind='bar', x='Name', y='Goals', title='Top 10 Goalscore in 2015/16', color='green' )
plt.show()
```

Trên đây là top 10 cầu thủ ghi nhiều bàn thắng nhất của mùa giải 2015/16. Thống kê cho thấy Harry Kane với số bàn thắng là 25

```python
#TOP 10 ASSIST IN 2015/16
assist = pl16.nlargest(10, 'Assists')
assist.plot(kind='bar', x='Name', y='Assists', title='Top 10 Assists in 2015/16')
plt.show()
```

Trên đây là top 10 cầu thủ kiến tại nhiều nhất của mùa giải 2015/16. Thống kê cho thấy Mesut Ozil với số kiến tạo là 18

```python
#TOP 10 INVOLVEMENTS IN 2015/16
pl16['involvements'] = pl16['Goals'] + pl16['Assists']
involvement = pl16.nlargest(10, 'involvements')
involvement.plot(kind='bar', x='Name', y='involvements', title='Top 10 involvements in 2015/16', color='brown')
plt.show()
```

Trên đây là top 10 cầu thủ đóng góp nhiều nhất vào các bàn thắng trong mùa giải 2015/16. Thống kê cho thấy Jamie Vardy đóng góp vào số bàn thắng của toàn đội nhiều nhất với 30 lần

```python
##Thống kê về các tiền vệ

mid16 = pl16[pl16['Position']=='Midfielder']

interception = mid16.nlargest(10, 'Interceptions')[['Name', 'Interceptions']]
foul = mid16.nlargest(10, 'Fouls')[['Name', 'Fouls']]
big_chances_created = mid16.nlargest(10, 'Big chances created')[['Name', 'Big chances created']]
#passes = mid16.nlargest(10, 'Passes')[['Name', 'Passes']]

for x in [interception, foul, big_chances_created]:
    print(x, end="\n")
```

## Trong tất cả các mùa

1. Top 10 cầu thủ ghi nhiều bàn thắng nhất



```python
top_scores = pl_upto_20.groupby('Name')['Goals'].sum()
top_10 = top_scores.nlargest(10)
top_10_plot = top_10.plot(kind='barh', x='Name', y='Goals', title='Top 10 Scores in 5 ss', )
print("{}\n\n{}".format(top_10, top_10_plot));
```

2. Những cầu thủ kiến tạo nhiều nhất

```python
top_assist = pl_upto_20.groupby('Name')['Assists'].sum()
top_10 = top_assist.nlargest(10)
top_10_plot = top_10.plot(kind='barh', x='Name', y='Assits', title='Top 10 Asstit in 5 ss')
print('{}\n\n{}'.format(top_10, top_10_plot));
```

3. Mùa nào có nhiều bàn thắng nhất

```python
pl16_goals = int(pl16['Goals'].sum())
pl17_goals = int(pl17['Goals'].sum())
pl18_goals = int(pl18['Goals'].sum())
pl19_goals = int(pl19['Goals'].sum())
pl20_goals = int(pl20['Goals'].sum())

pl16_assits = int(pl16['Assists'].sum())
pl17_assits = int(pl17['Assists'].sum())
pl18_assits = int(pl18['Assists'].sum())
pl19_assits = int(pl19['Assists'].sum())
pl20_assits = int(pl20['Assists'].sum())

overall = pd.DataFrame({
    'Year': ['2015/16', '2016/17', '2017/18', '2018/19', '2019/20'],
    'Goals': [pl16_goals, pl17_goals, pl18_goals, pl19_goals, pl20_goals],
    'Assists': [pl16_assits, pl17_assits, pl18_assits, pl19_assits, pl20_assits]
    
})

overall_plot = overall.plot(kind='line', x='Year', y='Goals', title='Overall Goals', marker='o')
plt.grid()
overall_plot_1 = overall.plot(kind='line', x='Year', y='Assists', title='Overall Assists', marker='o')
plt.grid()
print('{}\n\n\n{}\n\n\n{}'.format(overall, overall_plot, overall_plot_1));
```
